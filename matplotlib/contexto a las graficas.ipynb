```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = np.sin(x)
```

```python
fig, axes = plt.subplots(1,2,figsize=(6,5))
axes[0].plot(x,y,label="$sin(x)$")
axes[0].set_title('Relacion X - Y')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')
axes[0].legend()

axes[1].plot(y,x)
axes[1].set_title('Relacion Y - X')
axes[1].set_xlabel('Y')
axes[1].set_ylabel('X')
```

```python
plt.plot(x,y,label='sin(x)')
plt.title('Este es un titulo')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='lower left', bbox_to_anchor=(1,0.9))
```

```python
nrows=3
ncols=3
fig,axes=plt.subplots(nrows,ncols, figsize=(12,6))
for row in range(nrows):
    for col in range(ncols):
        axes[row,col].plot(x,np.sin(x))
        axes[row,col].set_title("sin(x)")
fig.tight_layout()
```

```python
y = np.zeros(100)
x = np.linspace(0,20,100)
n = np.linspace(1,9,9, endpoint=True)
harmonics = []
for idx, element in enumerate(n):
    y=np.sin(x*element)
    harmonics.append(y)

harmonics = np.array(harmonics)
harmonics = np.expand_dims(harmonics, axis=0)
harmonics = harmonics.reshape((3,3,100))
```

```python
nrows=3
ncols=3
fig,axes=plt.subplots(nrows,ncols)
for idx in range(nrows):
    for jdx in range(ncols):
        axes[idx,jdx].plot(x,harmonics[idx,jdx])
        axes[idx,jdx].set_title("Harmonic of sin(x)")
fig.tight_layout()
```
